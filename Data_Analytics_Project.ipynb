```python
#checking for missing values for each column int he dataset
df.isnull().sum()
```

```python
#filling up the missing values in review rating based on each cateogry
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
#checking for missing values after replacing it with the median
df.isnull().sum()
```

```python
#converting the column names lower case and replacing spaces with underscore(_)
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(" ","_")
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
#checking the column names
df.columns
```

```python
#creating a new column "age_group" to classify customers into different age groups
labels=["Young Adult","Adult","Middle Aged","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
```

```python
#displaying the age an age group column together
df[["age","age_group"]].head(15)
```

```python
#creating new column named "frequency_purchase_days" to show the frequency of purchases in number of days

#creating a dictionary to contain all the value types from the column "frequency of purchases"
frequency_mapping= {"Fortnightly":14,
                    "Weekly":7,
                    "Annually":365,
                    "Monthly":30,
                    "Quarterly":90,
                    "Bi-Weekly":14,
                    "Every 3 Months":90}
df["frequency_purchase_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
#displaying the new column with the old column "frequency_pf_purchases"
df[["frequency_of_purchases","frequency_purchase_days"]].head(20)
```

```python
#checking if the columns "promo_code_used" and "discount_applied" have the same value in each entry
#if same then dropping any one of the column
df[["discount_applied","promo_code_used"]].head(10)
(df["discount_applied"]==df["promo_code_used"]).all()
```

```python
#removing the column "promo_code_used" since the values of the two columns are same
df=df.drop('promo_code_used',axis=1)
#displaying the list of columns in the dataset
df.columns
```
